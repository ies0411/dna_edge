```python
import numpy as np
import os
import random
from natsort import natsorted
import shutil

MODE = "vehicle" #vehicle,edge

if MODE == "edge":
    pcd_path_prefix = "/mnt/nas3/Data/dna_autonomous/edge_ego/custom/points"
    imagesset_prefix = "/mnt/nas3/Data/dna_autonomous/edge_ego/custom/ImageSets"
    testset_path = "/mnt/nas3/Data/dna_autonomous/edge_ego/test/points"
elif MODE == "vehicle":
    pcd_path_prefix = "/mnt/nas3/Data/dna_autonomous/vehicle_ego/custom/points"
    imagesset_prefix = "/mnt/nas3/Data/dna_autonomous/vehicle_ego/custom/ImageSets"
    testset_path = "/mnt/nas3/Data/dna_autonomous/vehicle_ego/test/points"




train_val_prob = 50
```

```python


files = os.listdir(pcd_path_prefix)
files = natsorted(files)
npy_files = [f for f in files if f.endswith('.npy')]
print(f"Number of files: {len(npy_files)}")

# 51,234 : edge
# 26,693 : vehicle
```

```python


def move_file(source_path, destination_path):
    # Check if the source file exists
    if not os.path.exists(source_path):
        print(f"The file {source_path} does not exist.")
        return
    
    # Check if the destination directory exists, create if not
    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
    
    try:
        # Move the file
        shutil.move(source_path, destination_path)
        print(f"File moved to {destination_path}")
    except Exception as e:
        print(f"Error moving file: {e}")

```

```python
os.makedirs(imagesset_prefix, exist_ok=True)


with open(os.path.join(imagesset_prefix, "val.txt"), 'w') as val_file, open(os.path.join(imagesset_prefix, "train.txt"), 'w') as train_file:
    # for i in range(len(npy_files)):
    for idx, npy_file in enumerate(npy_files):
        # print(npy_file.split(".")[0][0])
        # break
        if npy_file.split(".")[0][0] == str(1):
            #copy
            move_file(os.path.join(pcd_path_prefix,npy_file), os.path.join(testset_path,npy_file))
            continue
        if idx%train_val_prob ==0:
            val_file.write(f'{npy_file.split(".")[0]}\n')
            # val_file.write(f'{str(i).zfill(8)}\n')
        else:
            train_file.write(f'{npy_file.split(".")[0]}\n')

            # train_file.write(f'{str(i).zfill(8)}\n')
```
